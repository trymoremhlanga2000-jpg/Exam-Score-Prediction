```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df.drop('student_id',axis=1,inplace=True)
```

```python
df.head()
```

```python
# Data Preprocessing
# Encodinh
```

```python
pd.get_dummies(df, columns=['gender','course','internet_access','study_method','facility_rating','exam_difficulty'],drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
columns = ['gender','course','internet_access','study_method','facility_rating','exam_difficulty','sleep_quality']
for col in columns:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
df.shape
```

```python
#Seprating input and output features
X = df.drop('exam_score',axis=1)
y = df['exam_score']
```

```python
X.shape
```

```python
# Train test split
# Standard scaler
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler


sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# making ANN Model
# Steps:-
#1 importing seque
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.activations import relu,linear
```

```python
model = Sequential()
```

```python
# input layer
model.add(Dense(units=11,activation='relu'))
```

```python
# Hidden Layer1
model.add(Dense(units=64,activation='relu'))
```

```python
# Hidden Layer2
model.add(Dense(units=32,activation='relu'))
model.add(Dropout(0.2))
```

```python
# Output layer
model.add(Dense(units=1,activation='linear'))
```

```python
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early = EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=0,
    verbose=0,
    mode='auto',
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0
)
```

```python
#Train the model
model.fit(X_train,y_train,epochs=100,batch_size=32,validation_split=0.3, callbacks=[early])
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = model.evaluate(X_test,y_test)
```

```python
# Evaluation
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
# R2 Score
r2 = r2_score(y_test, y_pred)
print("R Score:", r2)

# MSE
mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)

# MAE
mae = mean_absolute_error(y_test, y_pred)
print("MAE:", mae)

# RMSE
rmse = np.sqrt(mse)
print("RMSE:", rmse)
```
